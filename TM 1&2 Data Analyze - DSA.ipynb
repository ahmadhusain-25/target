# ANALISIS DESKRIPTIF

```python
# cek tipe data dan missing value
data_url.info()
```

```python
data_url['Order Date'] = pd.to_datetime(data_url['Order Date'], dayfirst=True, errors='coerce')
```

```python
# Clean numeric columns and calculate correlation
for col in ['Sales', 'Discount', 'Profit']:
    data_url[col] = data_url[col].astype(str).str.replace(',', '.').astype(float)

corr = data_url[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
print(corr)
```
